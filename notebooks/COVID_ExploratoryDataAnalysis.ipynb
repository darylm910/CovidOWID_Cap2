# COVID-19 Mortality Across Countries: An Exploratory Data Analysis

## Project Overview

The COVID-19 pandemic affected countries very differently. Some countries experienced relatively low mortality rates, while others suffered substantially higher death rates despite access to modern healthcare systems and vaccines.

The purpose of this analysis is to explore factors associated with differences in COVID-19 mortality across countries. To support this investigation, data were collected and merged from multiple sources, including:

- Our World in Data (OWID) COVID-19 statistics
- Oxford COVID Government Response Tracker Stringency Index
- World Bank demographic and healthcare indicators

The resulting dataset contains country-level information on COVID-19 cases, deaths, vaccination coverage, government response measures, demographics, healthcare resources, and economic conditions.

The primary outcome of interest is:

**Total COVID-19 deaths per million population**

This exploratory analysis will examine the distribution of mortality across countries, assess data quality, and investigate relationships between mortality and potential explanatory variables. The findings will help identify factors that may be useful for subsequent predictive modeling.
